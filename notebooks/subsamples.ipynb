```python
import matplotlib.pyplot as plt
from coffea.util import load
import hist
```

```python
filename = "/work/mmarcheg/BTVNanoCommissioning/output/test/fatjet_base_2018UL_v105/output.coffea"
output = load(filename)
output["variables"].keys()
```

```python
h = output["variables"]["JetGood_pt_1"]
h.keys()
```

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
slicing_data = {'cat' : 'inclusive', 'year' : '2018'}
slicing_mc   = {'cat' : 'inclusive', 'year' : '2018', 'variation' : 'nominal'}
for sample, histo in h.items():
    if sample == 'DATA':
        continue
        _h = histo[slicing_data]
    else:
        _h = histo[slicing_mc]
    _h.plot1d(stack=True, histtype='fill', ax=ax)
```
