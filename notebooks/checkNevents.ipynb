```python
import json
import matplotlib.pyplot as plt
```

```python
filename = "das.json"
with open(filename, 'r') as f:
    filelist = json.load(f)
```

```python
filelist.keys()
```

```python
Nevents = []
for record in filelist['data']:
    #print(record.keys())
    for file in record['file']:
        Nevents.append(file['nevents'])
Nevents
```

```python
plt.hist(Nevents, bins=1000, range=(0,10**6));
plt.xlabel("Nevents")
plt.ylabel("Nfiles")
plt.xlim(0, 10**5)
#plt.xscale("log");
```
