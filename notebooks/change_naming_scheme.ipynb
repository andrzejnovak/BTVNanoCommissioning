```python
import numpy as np
import pickle
```

```python
filename = "/work/mmarcheg/BTVNanoCommissioning/histograms/pileupJECfixed_2016EOY_wp064_pt450.pkl"
output = np.load(filename, allow_pickle=True)
```

```python
output.keys()
```

```python
output_new = {}
for key, value in output.items():
    if 'DDB' in key:
        output_new[key.replace('msd100tau06DDB', 'msd100tau06btagDDBvLV2')] = output[key]
    elif 'DDC' in key:
        output_new[key.replace('msd100tau06DDC', 'msd100tau06btagDDCvLV2')] = output[key]
        
output_new        
```

```python
outputFileName = filename.replace('.pkl', '_NewNamingScheme.pkl')
outputFile = open( outputFileName, 'wb' )
pickle.dump( output_new, outputFile, protocol=2 )
outputFile.close()
```
